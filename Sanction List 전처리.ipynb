```python
sdn = pd.read_csv('sanctionlist/sdn.CSV', header=None)
sdn.columns = ['ent_num', 'SDN_Name','SDN_Type','Program','Title','Call_Sign','Vess_type','Tonnage','GRT','Vess_flag','Vess_owner','Remarks']
```

```python
sdn['SDN']='Y'
```

```python
# 이름에서 null값이 들어간 행을 확인했다.
sdn.loc[sdn.SDN_Name.isnull(),:]
```

```python
sdn = sdn.loc[sdn.SDN_Name.isnull()!=True,:]
```

```python
non_sdn = pd.read_csv('sanctionlist/CONS_PRIM.CSV', header=None)
non_sdn.columns = ['ent_num', 'SDN_Name','SDN_Type','Program','Title','Call_Sign','Vess_type','Tonnage','GRT','Vess_flag','Vess_owner','Remarks']
```

```python
# sdn 구분칼럼을 만들어 넣어주었고
non_sdn['SDN']='N'
non_sdn = non_sdn.loc[non_sdn.SDN_Name.isnull()!=True,:]
```

```python
# 겹치는 이름 확인
for n in sdn.ent_num:
    if n in list(non_sdn.ent_num):
        print(n)
```

In order to make it easier to comply with OFAC's sanctions regulations,

These consolidated files comply with all OFAC's existing data standards.  

In the future, if OFAC creates a new non-SDN style list, 
the office will add the new data associated with that list to these consolidated data files if appropriate.
While the consolidated sanctions list data files are not part of OFAC's list of Specially Designated Nationals

### 겹치는 경우 non_SDN 항목에서 없애준다 

```python
for n in non_sdn.ent_num:
    if n in list(sdn.ent_num):
        print(n)
        
con = [n for n in non_sdn.ent_num if n not in list(sdn.ent_num)]
```

```python
non_sdn = non_sdn.loc[non_sdn.ent_num.isin(con)]
```

# Address table, text file name CONS_ADD.CSV

## Column

|sequence |Column name | Type   |  Size | Description |
|-------- |------------| ------- | ---- | ---------------------|
|1    |   Ent_num  |    number    |  |   link to unique listing |
|2    |    Add_num   |   number    |  |   unique record identifier |
|3    |    Address   |   text  |   750  | street address |
|4    |    City/State/Province/Postal Code		|	text   |  116 |  city, state/province, zip/postal code |
|5      |  Country    |  text   |  250  | country of address |
|6      |  Add_remarks | text   |  200  | remarks on address |

```python
add = pd.read_csv('sanctionlist/add.CSV', header=None)
add.columns = ['ent_num','Add_num','Address','City/State/Province/Postal Code','Country','Add_remarks']
add.info()
```

```python
cons_add = pd.read_csv('sanctionlist/CONS_ADD.CSV', header=None)
cons_add.columns = ['ent_num','Add_num','Address','City/State/Province/Postal Code','Country','Add_remarks']
cons_add.info()
```

### ent_num의 null 값 없애주기

```python
add.loc[add.Add_num.isnull(),:]
```

```python
cons_add.loc[cons_add.Add_num.isnull(),:]
```

```python
add = add.loc[add.Add_num.isnull()!=True,:]
cons_add = cons_add.loc[cons_add.Add_num.isnull()!=True,:]
```

### ent_num의 중복값 제거

```python
# 중복값 찾기
for n in add.ent_num:
    if n in list(cons_add.ent_num):
        print(n)
```

```python
con = [n for n in cons_add.ent_num if n not in list(add.ent_num)]
cons_add = cons_add.loc[cons_add.ent_num.isin(con)]
```

```python
cons_add
```

### 전처리 후 합치기

```python
add = pd.concat([add, cons_add], ignore_index=True)
add
```
